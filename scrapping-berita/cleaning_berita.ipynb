```python
df = df.drop_duplicates(subset=['Judul'], keep='first')

df
```

```python
# Buat fungsi untuk mengkategorikan berita
def kategorikan_berita(judul):
    # Gunakan regex untuk mencari kata kunci politik
    kata_kunci_politik = r"prabowo|israel|kepresidenan|caleg|sandi|gibran|cagub|dedi mulyadi|partai|budi arie|menkominfo|menkes|sri mulyani|KPU|PDIP|pemerintah|AHY|tom lembong|DPR|KPK|menko|jokowi|megawati|pilkada|pilgub|dpr|menteri|presiden|wapres|menteri|politik|kpk|tni|polri|partai|pemilu|kementerian|gubernur|bupati|wali kota|koalisi|reshuffle|kabinet|uu|bumn|bansos|dinasti|ridwan kamil|pramono|pilkada|anies|ahok|komisioner|polisi|SBY|PKS|Gerindra|Nasdem|Demokrat|korupsi|demo|demonstrasi|bansos"
    if re.search(kata_kunci_politik, judul, re.IGNORECASE):
        return 1  # Berita politik
    else:
        return 0  # Bukan berita politik

# Terapkan fungsi ke kolom 'Judul' untuk membuat kolom baru 'kelas_politik'
df['kelas_politik'] = df['Judul'].apply(kategorikan_berita)

# Tampilkan DataFrame dengan kolom baru
print(df)

# Simpan DataFrame ke file CSV baru (opsional)
df.to_csv("data real.csv", index=False)
```
